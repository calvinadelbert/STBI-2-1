```python
def clean_text(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df = pd.read_csv("BBC NEWS dataset.csv", encoding="latin-1")

df = df.dropna(how="any", axis=1)
df.columns = ['title','pubDate','guid','link','description']

df.head()
```

```python
df['message_len'] = df['description'].apply(lambda x: len(x.split(' ')))
df.head()
```

```python
df['message_clean'] = df['description'].apply(clean_text)
df.head()
```

```python
stop_words = stopwords.words('english')
more_stopwords = ['u', 'im', 'c']
stop_words = stop_words + more_stopwords

def remove_stopwords(text):
    text = ' '.join(word for word in text.split(' ') if word not in stop_words)
    return text
    
df['message_clean'] = df['message_clean'].apply(remove_stopwords)
df.head()
```

```python
stemmer = nltk.SnowballStemmer("english")

def stemm_text(text):
    text = ' '.join(stemmer.stem(word) for word in text.split(' '))
    return text
```

```python
df['message_clean'] = df['message_clean'].apply(stemm_text)
df.head()
```

```python
def preprocess_data(text):
    # Clean puntuation, urls, and so on
    text = clean_text(text)
    # Remove stopwords
    text = ' '.join(word for word in text.split(' ') if word not in stop_words)
    # Stemm all the words in the sentence
    text = ' '.join(stemmer.stem(word) for word in text.split(' '))
    
    return text
```

```python
df['message_clean'] = df['message_clean'].apply(preprocess_data)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['title'])

df['target_encoded'] = le.transform(df['title'])
df.head()
```
